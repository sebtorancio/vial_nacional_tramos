```python
import requests
import json

url = "https://sheets.googleapis.com/v4/spreadsheets/17AqjqeNvM4nG6cOUsUFKFaKXMiNmztYfzHIxeM9FcXk/values/tablavisible?key=AIzaSyCq2wEEKL9-6RmX-TkW23qJsrmnFHFf5tY&alt=json"

resp = requests.get(url)
datos_vialidad = resp.json()

print(datos_vialidad)
```

```python
print(json.dumps(datos_vialidad, indent=4, ensure_ascii=False))
```

```python
import pandas as pd

estado_rutas=pd.DataFrame(datos_vialidad['values'][2:], columns=datos_vialidad['values'][1])

print(estado_rutas)
```

```python
print(estado_rutas.columns)
```

```python
import unicodedata

def normalize(s):
    return ''.join(
        c for c in unicodedata.normalize('NFKD', s)
        if not unicodedata.combining(c)
    ).lower()

# detectar la columna que corresponde a "conocé más"
col_original = None
for c in estado_rutas.columns:
    if normalize(c) == "conoce mas":
        col_original = c
        break

# renombrarla
if col_original:
    estado_rutas = estado_rutas.rename(columns={col_original: "Conocemas"})
```

```python
print(estado_rutas.columns)
```

```python
estado_rutas_filtrado = estado_rutas[['Provincia', 'Ruta', 'Tramo', 'Calzada', 'Km', 'Conocemas']]
print(estado_rutas_filtrado)
```

```python
print(estado_rutas_filtrado.Conocemas[0])
```

```python
# Generar un DataFrame por cada ruta
rutas = {
    ruta: df.copy()
    for ruta, df in estado_rutas_filtrado.groupby("Ruta", sort=False)
}
```

```python
print(rutas['3'])
```

```python
# 1) Copiamos la ruta 3 y guardamos el índice original para ordenar
rn3 = rutas['3'].copy()
rn3["orig_idx"] = rn3.index     # <-- guarda el orden REAL

# 2) Orden deseado de provincias
orden_rn3 = [
    "Buenos Aires",
    "Río Negro",
    "Chubut",
    "Santa Cruz",
    "Tierra del Fuego, Antártida e Islas del Atlántico Sur"
]

# 3) Armamos los bloques por provincia SIN alterar el orden interno
partes = []
for prov in orden_rn3:
    bloque = rn3[rn3["Provincia"] == prov]
    if not bloque.empty:
        # ordenar SOLO por el índice original
        bloque = bloque.sort_values("orig_idx")
        partes.append(bloque)

# 4) Concatenar en el orden de provincias y eliminar la columna auxiliar
rn3_ordenada = pd.concat(partes, ignore_index=True).drop(columns=["orig_idx"])

# 5) Crear ID nuevo limpio (1,2,3,...)
rn3_ordenada.insert(0, "ID", range(1, len(rn3_ordenada) + 1))

print(rn3_ordenada)
```

```python
# Creamos un diccionario donde guardaremos las rutas
diccs = {}

# Recorremos todas las variables definidas en el entorno
for nombre, valor in list(globals().items()):
    if nombre.startswith("rn") and nombre.endswith("_ordenada"):
        diccs[nombre] = valor

with pd.ExcelWriter("rutas.xlsx", engine="openpyxl") as writer:
    for nombre, dicc in diccs.items():
        hoja = nombre.replace("_ordenada", "").upper()

        df = pd.DataFrame(dicc)

        # Agregar columnas nuevas (vacías)
        nuevas = ["Inicio_coords", "Inicio_desc", "Final_coords", "Final_desc", "Nota"]
        for col in nuevas:
            df[col] = ""

        # ---- REORDENAR ----
        # Colocarlas después de 'Km' y antes de 'Conocemas'
        cols = df.columns.tolist()

        # Ubicación de 'Km'
        if "Km" in cols:
            pos_km = cols.index("Km") + 1
        else:
            raise ValueError("No se encontró la columna 'Km' en la tabla.")

        # Remover las nuevas columnas de su posición final
        for col in nuevas:
            cols.remove(col)

        # Insertarlas después de Km manteniendo el orden
        for i, col in enumerate(nuevas):
            cols.insert(pos_km + i, col)

        # Reordenar DataFrame
        df = df[cols]

        # Exportar
        df.to_excel(writer, sheet_name=hoja, index=False)

```
